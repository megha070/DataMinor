```python
for country in Countries:
    countryfolder = os.path.join(Dir,country)
    for category in Categories:
        folder = os.path.join(countryfolder,category)
        for img in os.listdir(folder):
            img_path = os.path.join(folder,img)
            data.append(img_path)
data
```

```python
if not os.path.isdir("test"):
  os.mkdir("test")

if not os.path.isdir("train"):
  os.mkdir("train")
```

```python
for c in Categories:
  p=os.path.join("test",c)
  if not os.path.isdir(p):
    os.mkdir(p)


for c in Categories:
  p=os.path.join("train",c)
  if not os.path.isdir(p):
    os.mkdir(p)
```

```python
splitTrain = 0.8
testimg = 0
trainimg = 0
for country in Countries:
    countryfolder = os.path.join(Dir,country)
    for category in Categories:
        folder = os.path.join(countryfolder,category)
        imgs = os.listdir(folder)
#         for img in os.listdir(folder):
#             img_path = os.path.join(folder,img)
        split = int(0.8 * len(imgs))
        test = imgs[split:]
        train = imgs[:split]
        testimg = testimg+len(test)
        trainimg = trainimg+len(train)
        #Train
        for imgTrain in train:
            src=os.path.join(folder,imgTrain)
            dest=os.path.join("train/"+category,imgTrain)
            shutil.move(src,dest)
         #Test
        for imgTest in test:
            src=os.path.join(folder,imgTest)
            dest=os.path.join("test/"+category,imgTest)
            shutil.move(src,dest)
```
